# Stability of Planetary Orbital Configurations Klassifier

```python
from spock import spockClassifier
```

# Initialize our model. By default it currently loads our XGBoost summary features model

```python
model = spockClassifier()
```

# Make a planetary configuration

```python
import rebound
%matplotlib inline

sim = rebound.Simulation()
sim.add(m=1.)
sim.add(m=1.e-5, P=1., e=0.03, l=0.3)
sim.add(m=1.e-5, P=1.2, e=0.03, l=2.8)
sim.add(m=1.e-5, P=1.5, e=0.03, l=-0.5)
sim.move_to_com()
fig = rebound.OrbitPlot(sim)
```

# Get probability of stability

```python
model.predict(sim)
```
